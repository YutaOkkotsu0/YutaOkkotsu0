```python
import math
import pandas as pd

data = pd.read_csv('./Gender-Data.csv')
val_k = []
distance = []

def calcDistance(test):
    global distance
    percent_80 = int(0.8 * len(data)) 
    for p in range(percent_80):
        dis = format(math.sqrt((data['Height (in cm)'][p] - test[0])**2 + (data['Weight (in lbs)'][p] - test[1])**2 + (data['Hair length (in inches)'][p] - test[2])**2 ), '.2f')
        distance.append(dis)


def neighbors(k):
    global distance
    global val_k
    while(k != 0):
        for p in range(len(distance)):
            if(distance[p] != -1):
                minimum_val = distance[p]
                index = p
                break

        for q in range(len(distance)):
            if(float(distance[q]) < float(minimum_val) and float(distance[q]) > -1.00):
                minimum_val = distance[q]
                index = q
        val_k.append((minimum_val, index))
        distance[index] = -1
        k = k - 1


def classify():
    global data
    global val_k
    global distance
    
    
    Female = 0
    Male = 0
    
    
    for i in range(len(val_k)):
        if(data['Gender'][val_k[i][1]] == 'Female'):
            Female += 1
        elif(data['Gender'][val_k[i][1]] == 'Male'):

            Male += 1
    
    val_k.clear()
    distance.clear() 
    if(Male > Female):
        print("Test sample is Male")
    elif(Female > Male):
        print("Test sample is Female")
    

def knn(test, k):
    calcDistance(test)
    neighbors(k)
    classify()
        
percent_80 = int(0.8 * len(data))
for p in range(percent_80, len(data)):
    v1 = data['Height (in cm)'][p]
    v2 = data['Weight (in lbs)'][p]
    v3 = data['Hair length (in inches)'][p]
    test = (v1, v2, v3)
    print(test)
    k = 8
    knn(test, k)
```
