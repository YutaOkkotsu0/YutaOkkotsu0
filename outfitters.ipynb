```python
from bs4 import BeautifulSoup as Soup
import requests
import pandas as pd
import json
```

```python
url = "https://outfitters.com.pk/collections/men-sweatshirts" 
resp = requests.get(url) 
resp = resp.text 
```

```python
soup = Soup(resp, 'html.parser')
```

```python
divs= soup.findAll('div', {'class':'product-bottom'})
```

```python
products = []
for div in divs:
    title = div.find('a', {'class':'product-title'}).text.strip()
    price = div.find('span', {'class': 'money'}).text.strip()
    jsonObj = {"title":title, "price": price}
    products.append(jsonObj)
```

```python
products
```

```python
df = pd.DataFrame(products)
df
```
